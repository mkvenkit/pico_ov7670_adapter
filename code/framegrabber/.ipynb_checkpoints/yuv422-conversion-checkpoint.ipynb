```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def add_border(ax, border_width=2, border_color='white'):
    """
    Add border to the displayed image
    """
    # Get the current axis limits
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # Add rectangle patch
    rect = Rectangle((-0.5, -0.5), 
                    width=xlim[1]-xlim[0], 
                    height=ylim[1]-ylim[0],
                    fill=False,
                    color=border_color,
                    linewidth=border_width)
    ax.add_patch(rect)

def extract_y_channel(yuv_data, width=320, height=240, flip_vertical=False, flip_horizontal=False):
    """
    Extract just the Y channel from YUV422 data.
    """
    yuv = np.frombuffer(yuv_data, dtype=np.uint8)
    yuv = yuv.reshape((height, width//2, 4))
    
    y0 = yuv[:, :, 0]
    y1 = yuv[:, :, 2]
    
    y = np.zeros((height, width), dtype=np.uint8)
    y[:, 0::2] = y0
    y[:, 1::2] = y1
    
    if flip_vertical:
        y = np.flipud(y)
    if flip_horizontal:
        y = np.fliplr(y)
    
    return y

def display_y_channel(yuv_data, width=320, height=240, flip_vertical=False, flip_horizontal=False):
    """
    Display Y channel with border
    """
    y = extract_y_channel(yuv_data, width, height, flip_vertical, flip_horizontal)
    
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.imshow(y, cmap='gray')
    add_border(ax)
    ax.axis('off')
    ax.set_title('Y Channel (Luminance)')
    plt.show()
    
    return y

def yuv422_to_rgb(yuv_data, width=320, height=240, flip_vertical=False, flip_horizontal=False):
    """
    Convert YUV422 to RGB
    """
    yuv = np.frombuffer(yuv_data, dtype=np.uint8)
    yuv = yuv.reshape((height, width//2, 4))
    
    y0 = yuv[:, :, 0].astype(np.float32)
    u = yuv[:, :, 1].astype(np.float32)
    y1 = yuv[:, :, 2].astype(np.float32)
    v = yuv[:, :, 3].astype(np.float32)
    
    y = np.zeros((height, width), dtype=np.float32)
    y[:, 0::2] = y0
    y[:, 1::2] = y1
    
    u = np.repeat(u, 2, axis=1)
    v = np.repeat(v, 2, axis=1)
    
    if flip_vertical:
        y = np.flipud(y)
        u = np.flipud(u)
        v = np.flipud(v)
    
    if flip_horizontal:
        y = np.fliplr(y)
        u = np.fliplr(u)
        v = np.fliplr(v)
    
    y = (y - 16) * 1.164
    u = u - 128
    v = v - 128
    
    kr = 0.299
    kg = 0.587
    kb = 0.114
    
    r = y + (2 - 2 * kr) * v
    g = y - kb * (2 - 2 * kb) * u / kg - kr * (2 - 2 * kr) * v / kg
    b = y + (2 - 2 * kb) * u
    
    r = np.clip(r, 0, 255)
    g = np.clip(g, 0, 255)
    b = np.clip(b, 0, 255)
    
    rgb = np.stack([r, g, b], axis=2).astype(np.uint8)
    
    return rgb

def display_yuv422(yuv_data, width=320, height=240, flip_vertical=False, flip_horizontal=False, border_width=2, border_color='gray'):
    """
    Convert YUV422 data to RGB and display with border
    
    Args:
        yuv_data: Raw bytes in YUV422 format
        width: Image width
        height: Image height
        flip_vertical: If True, flip the image top to bottom
        flip_horizontal: If True, flip the image left to right
        border_width: Width of the border in pixels
        border_color: Color of the border
    """
    rgb = yuv422_to_rgb(yuv_data, width, height, flip_vertical, flip_horizontal)
    
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.imshow(rgb)
    add_border(ax, border_width, border_color)
    ax.axis('off')
    ax.set_title('RGB Image')
    plt.show()
    
    return rgb
```
